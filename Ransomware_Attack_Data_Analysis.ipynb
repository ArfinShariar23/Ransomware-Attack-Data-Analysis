```python
sector = df["sector"].value_counts()
sector
```

```python
plt.figure(figsize=(10,2))
plt.xticks(rotation=70,fontsize=10)
snb.set_palette("colorblind")
snb.barplot(x=sector.index, y=sector.values)
plt.xlabel("Sector")
plt.ylabel("Count")
plt.title("Counts for each sector")
```

We can see that, Tech, Healthcare, Government & Academic etc. are on high risk of Ransomware Attack.

```python
year = df["YEAR"].value_counts();
colors = snb.color_palette('RdYlGn_r')
plt.figure(figsize=(10,6))
snb.barplot(x=year.index,y=year,palette=colors)
plt.xticks(rotation=20,fontsize=10)
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Counts of attacks by year")
plt.show()
```

From the figure we understand that, 2020 and 2021 is the year where moost of the Ransomware attack happens. Also we understand from this analysis that Ransomware attack is raising day by day.

```python
#Analysis Months with the most attack.
attacks_by_months = df["month"].value_counts()
sorted_months = attacks_by_months.sort_values(ascending = False)
sorted_monthshead = sorted_months.head(12)
plt.figure(figsize=(10,6))
snb.barplot(x = sorted_monthshead.index, y=sorted_monthshead.values, palette="viridis")
plt.xlabel("Months")
plt.ylabel("Number of Attack")
plt.title("Months with the most attacks")
plt.tight_layout()
plt.show()
```

In the May month, Their is High risk of Ransomware attack.

```python
df["location"].value_counts()
```

```python
attacks_by_location = df['location'].value_counts()
top_locations = attacks_by_location.head(10)
plt.figure(figsize=(10, 6))
snb.barplot(x=top_locations.index, y=top_locations.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('country')
plt.ylabel('Number of Attacks')
plt.title('countries with the Most Attacks')
plt.show()
```

USA is at the High risk of Attack compare to other developed countries

```python
most_used_ransomware = df['Ransomware'].value_counts()
most_used_ransomware_head = most_used_ransomware.head(20)
plt.figure(figsize=(10, 6))
snb.barplot(x=most_used_ransomware_head.index, y=most_used_ransomware_head.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Ransomware Type')
plt.ylabel('Number of Occurrences')
plt.title('Top 20 Most Used Ransomware Types')
plt.tight_layout()
plt.show()
```

From the figure we can unnderstand that, Most of the ransomware attack reason remain Unknown. So, it is hard to find the reason behind the ransomware attack.
